```python
## Key Ideas You may find intuitive


# 1.Order of operations ( pemdas ) corresponds to parentheses,exponents,multiplication,division,addition,subtraction
# 2. A function can be continuous or non continuous and it is continuous where for every possible value of x there is a value of y.
# 3. A function is continuous but curvy , rather than linear and straight , we call it curvilinear function.
# 4. Find out why it is important to understand if a function is continuous or not with respect to data science

## EULERS NUMBER
# 5. In data science the default base for the log is Euler's number e
# In many applications where the base doesn't matter , we pick the one that results in the simplest derivative and that is euler's number e
# e=lim n→∞(1+ 1/n) ^ n whatever the n is it always converges to the value 2.7189
# e is mostly valid for compounding formulas ..there we will fing that we can rewrite the formula that contains e which is technically the (1+ 1/n) ^ n

## LIMITS  (let's call it behaviour)
# 1.We sometimes want to know the behaviour of a function like if x approaches 0 what happens to y for y= f(x)
# 2.You have seen e is the result of forever extending n into infinity for the function lim n→∞(1+ 1/n) ^ n
# 3. Figure out how we use limits in data science 


## DERIVATIVES (let's call it rate-of-change)
# 1.A derivative tells the slope of a fuction and it is useful to measure the rate of change at any point in the function
# 2. When the slope is 0, that means we are at the minimum or maximum of an output variable.
# 3. Partial derivatives...rate of change wirh single variable
# 4. we can calculate derivative using limits by considering the step size approaching to be zero
# 5. Chain rule states that for a given function y (with input variable x) composed into another function z (with input variable y) we can find derivative of z with respect to x by multiplying the two respective derivatives together


## Integrals (let's call it area under the curve for a given range)
# 1. We can calculate integrals with limits again by having that width of small piece of rectangles to be approaching but never zero
#∫ f(x)dx from a to b =lim n→∞ ∑ i=1 to n f(x i)Δx
```

```python
## Limits

from sympy import *

n= symbols('n')
f = (1+(1/n))**n
result = limit(f,n,oo)

print(result)
print(result.evalf())
```

```python
## Derivatives

def f(x):
    return x**2

def dx_f(x):
    return 2*x

slope_at_2 = dx_f(2.0)
print(slope_at_2)
```

```python
## Calculating partial derivatives

from sympy import *
from sympy.plotting import plot3d

x,y = symbols('x,y')
f= 2*x**3 + 3*y**3
dx_f = diff(f,x)
dy_f = diff(f,y)

print(dx_f)
print(dy_f)

plot3d(f)
```

```python
## calculating integrals

from sympy import *

x= symbols('x')
f= x**2 + 1
area = integrate( f, (x,0,1))
print(area)
```
